几个参数，truncation 可截断，padding 补全，max_length， return_tensors='pt'


```python
test_sentences = ['today is not that bad', 'Today is so bad']
```

```python
model_name = 'bert-base-uncased'
```

```python
!pip install transformers
```

```python
from transformers import AutoTokenizer,AutoModel
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)
```

```python
batch_input = tokenizer(test_sentences, max_length=32, truncation=True, padding=True, return_tensors='pt')
```

```python
print(batch_input)
```

```python
tokenizer(test_sentences[0])
```

```python
tokenizer.encode(test_sentences[0])
```

```python
tokenizer.tokenize(test_sentences[0])
```

```python
tokenizer.convert_tokens_to_ids(tokenizer.tokenize(test_sentences[0]))
```

```python
tokenizer.decode([101, 2651, 2003, 2025, 2008, 2919, 102])
```

```python
tokenizer.vocab
```

```python
tokenizer.special_tokens_map
```

```python
tokenizer.convert_tokens_to_ids([special for special in tokenizer.special_tokens_map.values()])
```

```python
tokenizer(test_sentences,max_length=32,truncation=True,padding="max_length", return_tensors='pt')
```
